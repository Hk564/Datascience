```python
Clients  = pd.read_csv('bank-full (1).csv', sep=';')
```

```python
Clients.head()
```

```python
Clients
```

```python
#removing NA values in the data set
Clients = Clients.dropna()
Clients.shape
```

```python
print(len(Clients[''].unique()))
```

```python
Clients.drop(["job"],inplace = True,axis = 1)
```

```python
X =Clients.iloc[:,:-1]
Y = Clients.iloc[:,15].astype('category')
```

```python
X = X.drop(columns =['marital'],axis =1 )
```

```python
classifier = LogisticRegression() #Logistic regression and fitting the model
classifier.fit(X,Y)
```

```python
y_pred = classifier.predict(X)
```

```python
y_pred_df = pd.DataFrame({'actual':Y,
                          'predicted_prob':y_pred})
```

```python
#confusion matrix to check the accuracy of the model
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
```

```python
print(confusion_matrix)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr, tpr, thresholds = roc_curve(Y,classifier.predict_proba(X)[:,1])
auc = roc_auc_score(Y,y_pred)
import matplotlib.pyplot as plt
plt.plot(fpr,tpr,color='blue',label ='logit model')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False positive rate')
plt.ylabel('True positive Rate')
plt.legend()
plt.show()
```
